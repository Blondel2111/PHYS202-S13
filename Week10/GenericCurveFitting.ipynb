Curve Fitting Example 1
-

```python
from scipy.optimize import curve_fit
```

```python
def func(x,a,b,c):
    return (a*x**2 + b*x + c)
```

```python
x = arange(0,6,1)
y = array([-1.78,4.09,8.85,17.9,26.1,35.2])
yerr=array([.46,.658,.528,1.34,1.09,.786])
```

```python
#Make an initial guess at the parameters
p0 = [1.0,3.0,-2.0]
#Call the curve fitter and ave it return the optimized parameters (popt)
popt,pcov=curve_fit(func,x,y,p0,yerr)
```

```python
#Compute the parameter uncertainties from the covariance matrix
punc=zeros(len(popt))
for i in arange(0,len(popt)):
    punc[i]=sqrt(pcov[i,i])
    
print 'optimal parameters: ',popt
print 'uncertainties of parameters: ',punc
```

```python
errorbar(x,y,yerr,fmt='r*')
fitresult=func(x,popt[0],popt[1],popt[2])
plot(x,fitresult,'b-')
```

Curve Fitting Example 2
-

```python
waveform_1=load('waveform_1.npy')
t=np.arange(len(waveform_1))
```

```python
plot(t,waveform_1)
xlabel('Time (ns)')
ylabel('Intensity (bins)')
```

```python
def model(t,B,A,mu,sigma):
    return B+A*exp(-((t-mu)/sigma)**2)
```

```python
x0=array([3.,30.,15.,1.])
```

```python
popt,pcov=curve_fit(model,t,waveform_1,x0)
plot(t,waveform_1,label='data')
plot(t,model(t,popt[0],popt[1],popt[2],popt[3]),label='fit',lw=2)
legend(loc='best')
```
