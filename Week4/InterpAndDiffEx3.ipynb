```python
def Temp(Resistance):
    """Takes a value of resistance[Ohms] and returns the equivalent temperature[Celsius]"""
    import numpy as np
    from scipy.interpolate import interp1d
    Ts=np.arange(39,9,-1)
    Rs=np.array([1.496e6,1.521e6,1.547e6,1.574e6,1.603e6,1.634e6,1.666e6,1.700e6,1.736e6,1.774e6,1.815e6,1.857e6,1.902e6,1.950e6,2.000e6,2.053e6,2.110e6,2.169e6,2.233e6,2.300e6,2.371e6,2.446e6,2.526e6,2.610e6,2.700e6,2.795e6,2.897e6,3.004e6,3.118e6,3.239e6])
    RInterp=interp1d(Rs,Ts)
    IntRes=RInterp(Resistance)
    return IntRes
```

```python
def finiteDifference(x,y):
    """Returns the centered finite difference derivative of y with respect to x"""
    dydx=zeros(y.shape,float)
    dydx[1:-1]=(y[2:]-y[:-2])/(x[2:]-x[:-2])
    dydx[0]=(y[1]-y[0])/(x[1]-x[0])
    dydx[-1]=(y[-1]-y[-2])/(x[-1]-x[-2])
    return dydx
```

(b)

```python
x=arange(0,10.1,.1)
y=x**5
plot(x[:-1],finiteDifference(x,y)[:-1],'r')
plot(x,5*x**4,'b')
```

We can tell our finiteDifference function is working because our approximate differentiation of $y=x^5 $ looks almost exactly like the actual exact differentiated function $y=4x^5$.

(c)

```python
x=arange(0,10.1,.001)
y=x**5
plot(x[:-1],(-5*x**4 + finiteDifference(x,y))[:-1],'r')
grid(True)
```

With a spacing of .001, our approximation is exact to begin with and loses accuracy as the values of x increases. At x=10, we are accurate to 3 significant figures.

(d)

```python
x=arange(0,10.1,.00031)
y=x**5
plot(x[:-1],(-5*x**4 + finiteDifference(x,y))[:-1],'r')
grid(True)
```

In order to be accurate to 4 significant figues at x=10, we need a spacing of .00031 or less between our x values.

```python
def fourPtFiniteDiff(x,y):
    """Returns the four point finite difference derivative of y with respect to x"""
    dydx=zeros(y.shape,float)
    dydx[2:-3]=(y[0:-5]-8.*y[1:-4]+8.*y[3:-2]-y[4:-1])/(12*(x[1:-4]-x[0:-5]))
    dydx[0]=(y[1]-y[0])/(x[1]-x[0])
    dydx[1]=(y[2]-y[1])/(x[2]-x[1])
    dydx[-1]=(y[-1]-y[-2])/(x[-1]-x[-2])
    dydx[-2]=(y[-2]-y[-3])/(x[-2]-x[-3])
    dydx[-3]=(y[-5]-8*y[-4]+8*y[-2]-y[-1])/(12*(x[-2]-x[-3]))
    return dydx
```

```python
x=arange(0,10,.1)
y=x**5
plot(x,fourPtFiniteDiff(x,y))
print fourPtFiniteDiff(x,y)[-3]
print y[-5]
print -8*y[-4]
print 8*y[-2]
print -1*y[-1]
```
