```python
def Temp(Resistance):
    """Takes a value of resistance[Ohms] and returns the equivalent temperature[Celsius]"""
    import numpy as np
    from scipy.interpolate import interp1d
    Ts=np.arange(39,9,-1)
    Rs=np.array([1.496e6,1.521e6,1.547e6,1.574e6,1.603e6,1.634e6,1.666e6,1.700e6,1.736e6,1.774e6,1.815e6,1.857e6,1.902e6,1.950e6,2.000e6,2.053e6,2.110e6,2.169e6,2.233e6,2.300e6,2.371e6,2.446e6,2.526e6,2.610e6,2.700e6,2.795e6,2.897e6,3.004e6,3.118e6,3.239e6])
    RInterp=interp1d(Rs,Ts)
    IntRes=RInterp(Resistance)
    return IntRes
```

(b)

```python
print Temp(1.496e6)
print Temp(1.574e6)
print Temp(2.000e6)
print Temp(2.526e6)
print Temp(2.700e6)
print Temp(3.239e6)
```

```python
Rx=np.arange(1.496e6,3.239e6,10)
plot(Temp(Rx),Rx,linewidth=2)
xlabel('Temperature [$^0$C]',fontsize=14)
ylabel('Resistance [M$\Omega$]',fontsize=14)
yticks(arange(1e6,4e6,.5e6),arange(1,4,.5))
annotate('R=$1.75 M\Omega$',xy=(Temp(1.75e6),1.75e6),xytext=(18,1.78e6),fontsize=15,arrowprops=dict(facecolor='black',width=1,headwidth=5))
annotate('T='+str(Temp(1.75e6))+'$^0C$',xy=(Temp(1.75e6),1.75e6),xytext=(30,2.2e6),fontsize=15,arrowprops=dict(facecolor='black',width=1,headwidth=5))
```

```python
def finiteDifference(x,y):
    """Returns the centered finite difference derivative of y with respect to x"""
    dydx=zeros(y.shape,float)
    dydx[1:-1]=(y[2:]-y[:-2])/(x[2:]-x[:-2])
    dydx[0]=(y[1]-y[0])/(x[1]-x[0])
    dydx[-1]=(y[-1]-y[-2])/(x[-1]-x[-2])
    return dydx
```

(b)

```python
x=arange(0,10.1,.1)
y=x**5
plot(x,finiteDifference(x,y),'r')
plot(x,5*x**4,'b')
```

We can tell our finiteDifference function is working because our approximate differentiation of $y=x^5 $ looks almost exactly like the actual differentiated function $y=4x^5$.
