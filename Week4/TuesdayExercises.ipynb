Electric dipole potential
-
$V(x,y)=\frac{kq}{\sqrt{x^2+(y-d)^2}}-\frac{kq}{\sqrt{x^2+(y+d)^2}}$

$k=\frac{1}{4\pi\epsilon_0}$

```python
dx=.06
dy=.06
x=arange(-2.,2.,dx)
y=arange(-2.,2.,dy)

X,Y=meshgrid(x,y)
print X.ndim
print Y.shape
print Y

k=1; q=1; d=1
V=(k*q/(X**2+(Y-d)**2)**(1/2.))-(k*q/(X**2+(Y+d)**2)**(1/2.))
print V
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=figure()
ax=fig.gca(projection='3d')
p=ax.plot_surface(X,Y,V,rstride=1,cstride=1,linewidth=0,cmap=cm.RdYlBu_r)
#help(Axes3D.plot_surface)
#help(figure)
cb=fig.colorbar(p,shrink=.5)
```

```python
#Representing 3d data as contour plots
levels=array([8,4,2,1,.5,.25,0,-.25,-.5,-1,-2,-4,-8])
contour(X,Y,V,levels,linewidths=2,cmap=cm.RdYlBu_r)
```

```python
def pointPotential(x,y,q,posx,posy):
    """Returns the potential for a point particle [meters][coulombs]"""
    k=8.987551787*10**9
    Vxy=k*q/sqrt((x-posx)**2+(y-posy)**2)
    return Vxy
```

```python
x=1.;y=1.
V11=pointPotential(x,y,1e-9,0.,0.)
print V11
```

```python
x,y=meshgrid(arange(-2.,2.,.06),arange(-2.,2.,.06))
V_xy=pointPotential(x,y,1e-9,0.,0.)
print V_xy
```

```python
def dipolePotential(x,y,q,d_x,d_y):
    """Returns the electric potential for a dipole [meters][coulombs]"""
    Vxy=pointPotential(x,y,q,d_x/2.,d_y/2.)-pointPotential(x,y,q,-d_x/2.,-d_y/2.)
    return Vxy
```

```python
dx=.06
dy=.06
x=arange(-2.,2.,dx)
y=arange(-2.,2.,dy)
X,Y=meshgrid(x,y)

V=dipolePotential(X,Y,1e9,1,0)

from mpl_toolkits.mplot3d import Axes3D
fig=figure()
ax=fig.gca(projection='3d')
p=ax.plot_surface(X,Y,V,rstride=1,cstride=1,linewidth=0,cmap=cm.rainbow)
cb=fig.colorbar(p,shrink=.5)
```
