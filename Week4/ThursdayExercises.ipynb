# Using Boolean Masks For Finding Data

```python
theta = arange(0.,2*pi,pi/100)
plot(theta,sin(theta))
```

```python
mask = (sin(theta)>0) & (sin(theta)<0.5)
#print mask
plot (theta,sin(theta),theta[mask],sin(theta[mask]))
```

```python
y = sin(theta[mask])
jj = (y==max(y))
#print jj
y[jj] = NaN
plot(theta,sin(theta))
plot(theta[mask],y,linewidth=3,color="red")
```

# Interpolation

```python
theta = arange(0.,2*pi+.01,pi/10)
mask = (theta == .5)
#print mask
```

```python
newTheta = .5; newSin = 0.
from scipy.interpolate import interp1d
#help(interp1d)
sinInterp=interp1d(theta,sin(theta))
newSin=sinInterp(newTheta)
print newSin
plot(theta,sin(theta),newTheta,newSin,'r+',ms=20,mew=2)
```

```python
thetaNew=arange(0.,2*pi+0.1,pi/30)
sinNew = interp1d(theta,sin(theta))(thetaNew)
plot(theta,sin(theta),thetaNew,sinNew,'r.',ms=8)
```

```python
plot(thetaNew,sinNew-sin(thetaNew))
ylabel("Difference between interpolated and exact values")
```

```python
sinNewC = interp1d(theta,sin(theta),kind='cubic')(thetaNew)
subplot(2,1,1)
plot(thetaNew,sinNew-sin(thetaNew), thetaNew,sinNewC-sin(thetaNew),'r')
subplot(2,1,2)
plot(thetaNew,sinNewC-sin(thetaNew),'r')
```

# Differentiation

```python
x = arange(0,10.01,.1)
y = x**5
```

```python
plot(x,y)
```

```python
dy = diff(y) #y(n+1)-y(n)
print len(y), len(dy)
dx = diff(x)
dydx=dy/dx
print len(dydx)
```

```python
dydx1=zeros(y.shape,float)
dydx1[:-1] = diff(y)/diff(x) #Slice from beginning to end-1
dydx1[-1] = (y[-1]-y[-2])/(x[-1]-x[-2]) #Backwards difference method to add value to last spot

dydx2=zeros(y.shape,float)
dydx2[1:-1]=(y[2:]-y[:-2])/(x[2:]-x[:-2])
dydx2[0]=(y[1]-y[0])/(x[1]-x[0])
dydx2[-1]=(y[-1]-y[-2])/(x[-1]-x[-2])
```

$y=x^5, \frac{dy}{dx}=5x^4$

```python
dydxExact = 5*x**4
```

```python
plot(x,dydx1,label='forward difference')
plot(x,dydx2,label='center difference')
plot(x,dydxExact,label='exact')
legend(loc='upper left')
```

```python
percentError1=100*abs(dydx1[1:] - dydxExact[1:])/dydxExact[1:]
percentError2=100*abs(dydx2[1:] - dydxExact[1:])/dydxExact[1:]
semilogy(x[1:],percentError1,label='forward difference')
semilogy(x[1:],percentError2,label='center difference')
ylabel("percent error with dydxExact")
legend(loc="upper right")
```
