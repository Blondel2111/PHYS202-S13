Example 1: Analytic Inverse CDF Transform
-

```python
x=arange(0,pi+.01,.01)
f=lambda x:sin(x)/2
F=lambda x:(cos(0)-cos(x))/2

plot(x,f(x),label='True Distribution',lw=3)
plot(x,F(x),label='Cumulative function',lw=3)

xlim(0,3.5)
ylim(0,1.1)
legend(loc=0)
show()
```

```python
#Number of samples
N=10000
#N random numbers between 0 and 1
z=numpy.random.random_sample(N);

#inverse CDF
Finv=lambda z:arccos(1-2*z)
#sampled x values
xsamples=Finv(z);

#how do they compare to the original?
hist(xsamples,50,normed=True,label='Sampled Distribution')
plot(x,f(x),label='True Distribution',lw=3,color='r')
xlim(0,pi)
ylim(0,.7)
legend(loc=0)
```

Exercise 1. Exponential function - analytic
-

(a)

```python
lam=1
f=lambda x:lam*e**-(lam*x)
F=lambda x:1-1*e**-(lam*x)
Finv=lambda z:-log(1-z)/lam
```

(b)

```python
x=arange(0,5.01,.01)
plot(x,f(x),label='True Distribution',lw=3)
plot(x,F(x),label='Cumulative Distribution',lw=3)
legend(loc=5)
```

(c)

```python
N=10000
z=numpy.random.random_sample(N);
xsamples=Finv(z)
hist(xsamples,50,normed=True,label='Sampled distribution')
plot(x,f(x),label='True distribution',lw=3,color='red')

xlim(0,5)
ylim(0,.8)
legend()
```

Exercise 2. Exponential function - numerical
-

(a)

```python
#help(searchsorted)
#help(cumsum)
xmin=0
xmax=5
dx=.01
x=arange(xmin,xmax+dx,dx)
y=e**(-x)
A=cumsum(y)/(cumsum(y)[(xmax/dx)])
N=10000
z=numpy.random.random_sample(N)
B=searchsorted(A,z)
x[B]
```

(b)

```python
plot(x,y,label='True Distribution',lw=3)
plot(x,A,label='Cumulative Distribution',lw=3)
hist(x[B],50,normed=True,label='Sampled distribution')
legend(loc=5)
```

(c)

```python
def distribute1D(x,prob,N):
    """Takes any numerical distribution prob, on the interval defined by the array x, and returns an array of N sampled values that are statistically the same as the input data."""
    y=prob
    A=cumsum(y)/(cumsum(y)[(len(x)-1)])
    z=numpy.random.random_sample(N)
    B=searchsorted(A,z)
    return x[B]
```

```python
d=float(15e-6)
a=float(10e-6)
lam=632.8e-9
L=1.
x=arange(-.2,.2,.001)
N=10000.
y=(sin(pi*a*x/(lam*L))/(pi*a*x/(lam*L)))**2*cos(2*pi*d*x/(lam*L))**2
dist=distribute1D(x,y,N)
```

```python

n,bins,patches=hist(dist,50,normed=True,label='Sampled Distribution')
plot(x,y*max(n),lw=2,color='red',label='True Distribution')
xlim(-.2,.25)
legend(loc=1)
```
