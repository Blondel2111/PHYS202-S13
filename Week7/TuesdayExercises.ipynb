```python
from scipy.integrate import odeint
#odeint?
```

```python
def deriv(theta_current,t):
    g=9.8#m/s^2
    l=.5 #m
    theta=theta_current[0]
    omega=theta_current[1]
    dtheta=omega
    domega=-(g/l)*sin(theta)
    return array([dtheta,domega])
```

```python
t=arange(0.,10.01,.01)

omega0=0.
theta_init1=array([0.,omega0])
theta1 = odeint(deriv,theta_init1,t)

theta_init2=array([pi/4.,omega0])
theta2 = odeint(deriv,theta_init2,t)

theta_init3=array([pi/2.,omega0])
theta3 = odeint(deriv,theta_init3,t)

theta_init4=array([3*pi/4.,omega0])
theta4 = odeint(deriv,theta_init4,t)

theta_init5=array([pi,omega0])
theta5 = odeint(deriv,theta_init5,t)
```

```python
plot(t,theta1[:,0],'b')
plot(t,theta2[:,0],'r')
plot(t,theta3[:,0],'g')
plot(t,theta4[:,0],'black')
plot(t,theta5[:,0],'purple')
```

```python
l=.5
g=9.8
t=arange(0.,10.01,.01)
#theta_app=lambda t: th0*cos(sqrt(g/l)*t)
plot(t,0*cos(sqrt(g/l)*t))
plot(t,pi/4.*cos(sqrt(g/l)*t))
plot(t,pi/2.*cos(sqrt(g/l)*t))
plot(t,3*pi/4.*cos(sqrt(g/l)*t))
plot(t,pi*cos(sqrt(g/l)*t))
```

For the numerical solutions, the period changes as the initial angle increases. At th0=0, there is no oscillation. At th0=pi, the graph does not oscillate until a long amount of time.
