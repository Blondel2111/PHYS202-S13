```python
def is_anagram(s1,s2):
    a=s1.upper()
    b=s2.upper()
    L1=[]
    L2=[]
    for l in a:
        L1.append(l)
    for k in b:
        L2.append(k)
    G1=sort(L1)
    G2=sort(L2)
    for m in range (len(G1)):
        if G1[m]!=G2[m]:
            return False
            break
        else:
            return True
```

```python
print is_anagram('Tac','Cat')
```
