Using Matplotlib to make graphs

```python
#simple line plot
t = arange(0.0,1.01,.01)
c = cos(4*pi*t)
s = sin(4*pi*t)
```

```python
plot(t,s,color="blue",linewidth=2,label="$sin(4\pi\t)$")
plot(t,c,color="red",linewidth=2,label="$cos(4\pi\t)$")
xlabel("Time [s]",fontsize=14)
ylabel("Voltage [mV]",fontsize=14)
title("My cool plot",fontsize=20)
legend()
grid(True)
savefig("simple_plot.pdf")
```
