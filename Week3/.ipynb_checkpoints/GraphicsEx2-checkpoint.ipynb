```python
i=complex(0,1)
x=arange(-pi,pi,.01)
U=-1*cosh(x)**-2
k00=-1*tanh(x)
k00c=conjugate(k00)
k1=k00*k00c
k05=(.5*i-tanh(x))/(i*.5+1)*e**(i*.5*x)
k05c=conjugate(k05)
k2=k05*k05c
k10=(1.0*i-tanh(x))/(i+1)*e**(i*x)
k10c=conjugate(k10)
k3=k10*k10c
k20=(2.0*i-tanh(x))/(2.0*i+1)*e**(i*2.0*x)
k20c=conjugate(k20)
k4=k20*k20c
```

```python
plot(x,U,color="blue",linewidth=2,label="$U(x)$")
plot(x,k1,color="red",linewidth=2,label="$\psi^2, k=0.0$")
plot(x,k2,color="red",linewidth=2,linestyle="--",label="$\psi^2, k=0.5$")
plot(x,k3,color="red",linewidth=2,linestyle="-.",label="$\psi^2, k=1.0$")
plot(x,k4,color="red",linewidth=2,linestyle=":",label="$\psi^2, k=2.0$")
xlabel("$x(rad)$",fontsize=16)
ylabel("$U(x),\psi^2(x)$",fontsize=16)
grid(True)
xticks([-pi,-pi/2,0,pi/2,pi],('$-\pi$','$-{\pi}/{2}$','$0$','$+{\pi}/{2}$','$+\pi$'),fontsize=16)
yticks(arange(-1,1.5,.5),fontsize=11)
ylim(-1.2,1.2)
xlim(-pi-.5,pi+.5)
legend(loc=3,labelspacing=-.01,borderpad=0, columnspacing=0)
```

```python
data = loadtxt('/nfshome/crexrode/sunspots.txt',float)
x=data[:,0]
y=data[:,1]
```

```python
plot(x,y,ls=' ',marker='.',ms=1.5,mfc="black",mec="black")
xlabel("Year",fontsize=16)
ylabel("Number of sunspots Observed",fontsize=15)
xticks(arange(0,3142,240),1749+arange(0,262,20),rotation=-45)
grid(True)
```

(b)

```python
plot(x,y,ls=' ',marker='.',ms=1.5,mfc="black",mec="black")
xlabel("Year",fontsize=16)
ylabel("Number of sunspots Observed",fontsize=15)
xticks(arange(0,3142,120),1749+arange(0,262,10),rotation=-45)
grid(True)
xlim(0,1000)
```

(c)

```python
y_avg=[]
y_tot=0
for i in x:
    y_tot+=y[i]
    y_avg.append(y_tot/int(i))
```

```python
plot(x,y,ls='',marker='.',ms=1.5,mfc="black",mec="black",label="Quantity observed")
plot(x,y_avg,label="Running average")
xlabel("Year",fontsize=16)
ylabel("Number of sunspots Observed",fontsize=15)
xticks(arange(0,3142,120),1749+arange(0,262,10),rotation=-45)
grid(True)
xlim(0,1000)
legend()
```
